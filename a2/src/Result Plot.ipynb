```python
import pandas as pd

df = pd.read_csv("../results/throughput.csv")
df['AP1_MBytes'] = df['AP1_Bytes'] / (1024 * 1024)
df['AP1_MBytes'] = df['AP1_MBytes'].round(3)
df['AP2_MBytes'] = df['AP2_Bytes'] / (1024 * 1024)
df['AP2_MBytes'] = df['AP2_MBytes'].round(3)
df['throughput1'] = (df['AP1_MBytes'].diff() / df['Time'].diff()).fillna(0)
df['throughput2'] = (df['AP2_MBytes'].diff() / df['Time'].diff()).fillna(0)
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(10, 6))
plt.plot(df['Time'].values, df['throughput1'].values, label='AP Network A', color='blue')
plt.plot(df['Time'].values, df['throughput2'].values, label='AP Network B', color='red')
plt.xlabel('Time')
plt.ylabel('Mega Bytes')
plt.title('AP Network A MBytes vs AP Network B MBytes over Time')
plt.legend()
plt.grid(True)
plt.savefig('../results/throughput_plot.jpg', dpi=300)
plt.show()
```
